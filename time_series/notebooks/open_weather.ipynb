```python
import json
import pandas as pd
import os

df = pd.DataFrame(columns=['dt','main','wind','clouds','weather','rain','data','id'])

for d,r,files in os.walk('../open_weather/'):
    for file in files:
        with open(f'{d}{file}','r') as f:
            data = json.loads(f.read())
        c = pd.DataFrame(data['list'])
        c['id'] = file.split('.')[0]
        df = pd.concat([df,c])
df.head(10)
```

```python
df['data'] = pd.to_datetime(df['dt'], unit='s')
df.head(10)
```

```python
df['id'] = df['id'].astype(int)
df.dtypes
```

```python
from cassandra.cluster import Cluster
import uuid
cluster = Cluster(['localhost'])
session = cluster.connect('prediction')
for i in df.iterrows():
    query = "INSERT INTO data_series (data_id, id_municipio, temp, pressure, humidity, wind_speed, clouds, data) VALUES (%s,%s, %s, %s,%s,%s,%s,%s)"
    session.execute(
        query, (
            uuid.uuid4(), 
            i[1]['id'], 
            i[1]['main']['temp'],
            i[1]['main']['pressure'],
            i[1]['main']['humidity'], 
            i[1]['wind']['speed'],
            i[1]['clouds']['all'],
            f'{i[1]["data"].strftime("%Y-%m-%d %H:%M:%S")}'))
session.shutdown()
cluster.shutdown()
```
